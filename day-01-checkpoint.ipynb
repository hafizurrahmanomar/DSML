````python
print(2 + 3)   # addition(+)
print(3 - 1)   # subtraction(-)
print(2 * 3)   # multiplication(*)
print(3 / 2)   # division(/)
print(4 ** 2)  # exponential(**)
print(5 % 2)   # modulus(%)
print(9 // 2)  # Floor division operator(//)
````
